### math & round

````python
'''
讓使用者輸入直角三角形的對邊
讓使用者輸入直角三角形的斜邊
計算角度
公式:
sin(x) = 對邊 / 斜邊
x是radian(弧度)
'''

import math

opposite = float(input("請輸入直角三角形的對邊："))
hypotenuse = float(input("請輸入直角三角形的斜邊："))

sin_x = opposite / hypotenuse
x_radian = math.asin(sin_x)

# 使用round()進行四捨五入
print("角度(弧度)=", round(x_radian, 2))
print("角度(度數)=", round(math.degrees(x_radian), 2))
````

### 巢狀判斷

````python
#優化上題
#學生總分為300
#有些學生可以加分5%
#如果加分超過300，就以300分為準

#使用 int() 將輸入的字串轉換為整數
scores:int|float = int(input("請輸入學生分數(最高300分)"))
if scores <= 300:
  is_add:str = input("學生是否符合加分條件?(y,n)")

  if is_add == "y":
    scores = scores * 1.05
    if scores > 300: #巢狀判斷
      scores = 300
  print("學生分數是：", round(scores ,2))
else:
  print("輸入的分數,不可以大於300")
````

````python
#根據上題做其它延伸練習
scores:int|float = int(input("請輸入學生分數(最高300分)"))
if scores <= 300:
  if scores < 200:
    scores = scores * 1.05
    print(scores)
  elif scores >= 200: # 修正：else 不能加條件，應改用 elif: 如果前面的條件不成立，則檢查這個新條件
    scores = scores
    print(scores)
else:
  scores = 300
  print(scores)
````

### elif 的定義與用法

`elif` 是 **else if** 的縮寫。它用於多重條件判斷的流程中：

1.  **if**: 檢查第一個條件。如果成立，執行該區塊並結束判斷。
2.  **elif**: 如果前面的條件**不成立**，則檢查這個新條件。可以有多個 `elif`。
3.  **else**: 如果前面的 `if` 和所有 `elif` 條件都**不成立**，則執行這個「最後保底」的區塊。

**語法結構：**
```python
if 條件A:
    # 執行 A
elif 條件B:
    # 執行 B
else:
    # 以上都不成立時執行
```

### `if-elif` vs 多個 `if` 的差異

假設我們要根據分數給評語：

#### 1. 使用 `if-elif` (正確做法)
一旦符合一個條件，後面的就不會執行。
```python
score = 90
if score >= 80:
    print("優秀")
elif score >= 60:
    print("及格")
# 結果：只會印出 "優秀"
```

#### 2. 全部改用 `if` (可能出錯)
程式會檢查「每一行」if，導致符合多個條件。
```python
score = 90
if score >= 80:
    print("優秀")
if score >= 60:
    print("及格")
# 結果：會同時印出 "優秀" 和 "及格"
```
